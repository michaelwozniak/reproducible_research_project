#### dividing covid data into 4 dataframes
- obtain two extreme dates
- difference in total cases
- dividing total cases by population (multiplied by 100,000)

```python
covid_data_Q2 = covid_data[covid_data.date.isin(['2020-02-28', '2020-06-30'])]
covid_data_Q2['covid_cases'] = covid_data_Q2.groupby(by=["location"])['total_cases'].diff()
covid_data_Q2 = covid_data_Q2[covid_data_Q2.date.isin(['2020-06-30'])]
covid_data_Q2['cases_ratio'] = covid_data_Q2['covid_cases'] / covid_data_Q2['population']*100000
covid_data_Q2.loc[covid_data_Q2['cases_ratio'].isna() == True, ['cases_ratio']] = covid_data_Q2['total_cases']/covid_data_Q2['population']
covid_data_Q2 = covid_data_Q2[['iso_code', 'location', 'cases_ratio']]

covid_data_Q3 = covid_data[covid_data.date.isin(['2020-06-30', '2020-09-30'])]
covid_data_Q3['covid_cases'] = covid_data_Q3.groupby(by=["location"])['total_cases'].diff()
covid_data_Q3 = covid_data_Q3[covid_data_Q3.date.isin(['2020-09-30'])]
covid_data_Q3['cases_ratio'] = covid_data_Q3['covid_cases'] / covid_data_Q3['population']*100000
covid_data_Q3.loc[covid_data_Q3['cases_ratio'].isna() == True, ['cases_ratio']] = covid_data_Q3['total_cases']/covid_data_Q3['population']
covid_data_Q3 = covid_data_Q3[['iso_code', 'location', 'cases_ratio']]

covid_data_Q4 = covid_data[covid_data.date.isin(['2020-09-30', '2020-12-31'])]
covid_data_Q4['covid_cases'] = covid_data_Q4.groupby(by=["location"])['total_cases'].diff()
covid_data_Q4 = covid_data_Q4[covid_data_Q4.date.isin(['2020-12-31'])]
covid_data_Q4['cases_ratio'] = covid_data_Q4['covid_cases'] / covid_data_Q4['population']*100000
covid_data_Q4.loc[covid_data_Q4['cases_ratio'].isna() == True, ['cases_ratio']] = covid_data_Q4['total_cases']/covid_data_Q4['population']
covid_data_Q4 = covid_data_Q4[['iso_code', 'location', 'cases_ratio']]

covid_data_all = covid_data[covid_data.date.isin(['2020-02-28', '2020-12-31'])]
covid_data_all['covid_cases'] = covid_data_all.groupby(by=["location"])['total_cases'].diff()
covid_data_all = covid_data_all[covid_data_all.date.isin(['2020-12-31'])]
covid_data_all['cases_ratio'] = covid_data_all['covid_cases'] / covid_data_all['population']*100000
covid_data_all.loc[covid_data_all['cases_ratio'].isna() == True, ['cases_ratio']] = covid_data_all['total_cases']/covid_data_all['population']
covid_data_all = covid_data_all[['iso_code', 'location', 'cases_ratio']]
```

### Final merging

#### Imputation for Turkmenistan and North Korea
- Turkmenistan 0 confirmed cases according to WHO https://covid19.who.int/region/euro/country/tm
- North Korea 0 confirmed cases according to WHO https://covid19.who.int/region/searo/country/kp

- imputation - mean value of Asian countries

```python
data_Q2.groupby('continent')['cases_ratio'].mean()['Asia']
```

```python
data_Q2.loc[data_Q2.country_name == 'North Korea', 'cases_ratio'] = data_Q2.groupby('continent')['cases_ratio'].mean()['Asia']
data_Q2.loc[data_Q2.country_name == 'Turkmenistan', 'cases_ratio'] = data_Q2.groupby('continent')['cases_ratio'].mean()['Asia']
data_Q2.drop(columns = {'iso_code', 'location'}, inplace = True)

data_Q3.loc[data_Q3.country_name == 'North Korea', 'cases_ratio'] = data_Q3.groupby('continent')['cases_ratio'].mean()['Asia']
data_Q3.loc[data_Q3.country_name == 'Turkmenistan', 'cases_ratio'] = data_Q3.groupby('continent')['cases_ratio'].mean()['Asia']
data_Q3.drop(columns = {'iso_code', 'location'}, inplace = True)

data_Q4.loc[data_Q4.country_name == 'North Korea', 'cases_ratio'] = data_Q4.groupby('continent')['cases_ratio'].mean()['Asia']
data_Q4.loc[data_Q4.country_name == 'Turkmenistan', 'cases_ratio'] = data_Q4.groupby('continent')['cases_ratio'].mean()['Asia']
data_Q4.drop(columns = {'iso_code', 'location'}, inplace = True)

data_all.loc[data_all.country_name == 'North Korea', 'cases_ratio'] = data_all.groupby('continent')['cases_ratio'].mean()['Asia']
data_all.loc[data_all.country_name == 'Turkmenistan', 'cases_ratio'] = data_all.groupby('continent')['cases_ratio'].mean()['Asia']
data_all.drop(columns = {'iso_code', 'location'}, inplace = True)
```
